```python
import pandas as pd
from sqlalchemy import create_engine
import numpy as np
```

```python
covid_states = "Resources/states.csv"
covid_states_df = pd.read_csv(covid_states)
covid_states_df
```

```python
#Cleaning rows to clean table
del covid_states_df['country']
del covid_states_df['county']
del covid_states_df['level']
del covid_states_df['lat']
del covid_states_df['locationId']
del covid_states_df['long']
del covid_states_df['url']
del covid_states_df['cdcTransmissionLevel']



covid_states_df.head()
```

```python
#Seperating the Data
covid_states_df.columns
```

```python
vaccine_distr_df = covid_states_df[["state", "population","actuals.vaccinesDistributed","actuals.vaccinationsCompleted"]].copy()
vaccine_distr_df = vaccine_distr_df.drop([25])
vaccine_distr_df = vaccine_distr_df.drop([40])
vaccine_distr_df = vaccine_distr_df.drop([7])
vaccine_distr_df = vaccine_distr_df.reset_index()
del vaccine_distr_df['index']


vaccine_distr_df = vaccine_distr_df.rename(columns={"state":"State",
                                                                "population":"Population",
                                                                "actuals.vaccinesDistributed":"Vaccines_distributed",
                                                                "actuals.vaccinationsCompleted":"Vaccines_completed"
                                                               })
vaccine_distr_df["Vaccine_perc"] = vaccine_distr_df["Vaccines_completed"].div(vaccine_distr_df["Vaccines_distributed"].values/100)

vaccine_distr_df['Vaccine_perc']= round(vaccine_distr_df['Vaccine_perc'],1)


vaccine_distr_df
```

```python
hospital_df = covid_states_df[["state", "population", "actuals.hospitalBeds.capacity", "actuals.hospitalBeds.currentUsageTotal","actuals.hospitalBeds.typicalUsageRate"]].copy()
hospital_df = hospital_df.drop([25])
hospital_df = hospital_df.drop([40])
hospital_df = hospital_df.drop([7])
hospital_df = hospital_df.reset_index()
del hospital_df['index']


hospital_df = hospital_df.rename(columns={"state":"State",
                                                    "population":"Population",
                                                    "actuals.hospitalBeds.capacity":"Bed_capacity",
                                                    "actuals.hospitalBeds.currentUsageTotal":"Bed_current_capacity",
                                                    "actuals.hospitalBeds.typicalUsageRate":"Typical_usage_rate"
                                                    })

hospital_df["Typical_usage_rate"] = hospital_df["Typical_usage_rate"].values*100


hospital_df
```

```python
hospital_df.to_csv("Data/hospital.csv",index=False)
vaccine_distr_df.to_csv("Data/vaccine_distr.csv",index=False)
```
